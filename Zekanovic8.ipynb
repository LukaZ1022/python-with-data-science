```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('avocado.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
filtered = df.query("year == 2015 and type == 'conventional' and region == 'Albany'")
df2 = filtered[['Date', 'Total Bags', 'Small Bags']]
df2.head()
```

```python
melted = pd.melt(df, id_vars='Date', value_vars=['Total Bags', 'Small Bags'], var_name='Bags', value_name='Count')
melted
```

```python
sns.relplot(melted, kind='line', x='Date', y='Count', hue='Bags')
```

```python
df['Sales Volume'] = pd.qcut(df['Total Volume'], q=4, labels=['Poor', 'Modest', 'Good', 'Excellent'])
df.head()
```

```python
sns.countplot(data=df, x='year', hue='Sales Volume')
```

```python
bag_melted = pd.melt(df, id_vars= 'year', value_vars = ['Small Bags', 'Large Bags', 'XLarge Bags'], var_name = 'Bags', value_name= 'Count')
bag_melted
```

```python
sns.relplot(data = bag_melted, kind='line', x='year', y='Count', hue='Bags')
```
