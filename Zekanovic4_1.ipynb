# Exercise 3-1: Create some plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Get the data

```python
mortality_data = pd.read_pickle('mortality_prepped.pkl')
mortality_data.head()
```

```python
mortality_wide = pd.read_pickle('mortality_wide.pkl')
mortality_wide.head()
```

## Visualize the data

```python
mortality_data.query('AgeGroup == "01-04 Years"').plot.scatter(x='Year', y='DeathRate')
```

```python
mortality_wide.plot()
```

```python
mortality_wide.plot.line(y='15-19 Years', title='Mortality Rates for the ages 15-19 in the wide data')
plt.legend().remove()
```

```python
mortality_wide.plot.area(legend='reverse')
```

```python
mortality_wide.query('Year == 1900').plot.bar()
```

```python
mortality_wide.query('Year == 2000').plot.bar(title='Mortality Rates for the year 2000 in the Mortality Wide Data', xlabel="")
```

```python
mortality_data.query('Year == 2000').pivot(index='AgeGroup', values='DeathRate', columns='Year').plot.bar(title='Long Data Mortality Rates for the year 2000', rot=45)
plt.legend().remove()
```

```python
mortality_data.plot.hist(bins = 15, y='DeathRate')
```

```python
mortality_data.query('Year == 1900').plot.kde(grid=True, title='Mortality Rates for the Year 1900', y='DeathRate')
```

```python
mortality_data.query('Year in (1900, 1925, 1950, 1975, 2000)').groupby('Year').DeathRate.sum().plot.pie()
```

```python
mortality_wide.query('Year in (1900, 1925, 1950, 1975, 2000)').plot.barh(title=['Mortality: Age 1-4', 'Mortality: Age 5-9','Mortality: 10-14', 'Mortality: Age 15-19'],
    sharey=True,subplots=True, layout=(2, 2),)
plt.legend().remove()
                                                                                
```
